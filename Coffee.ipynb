```python
#Готовим новый датафрейм
new_df=pd.DataFrame({})
new_df_date=[]
new_df_promo=[] 
new_df_trans=[]
new_df_negative=[]
new_df_income=[]
```

```python
for i in no_delivery.Date.unique():
    for j in (0,1):
        try:
            df_with_date=no_delivery.loc[(no_delivery['Date']==i)&(no_delivery['Promo']==j)]
        except:
            j-=1
            df_with_date=no_delivery.loc[(no_delivery['Date']==i)&(no_delivery['Promo']==(0,1)[j])]
        if df_with_date.empty is False:
            new_df_date.append(str(i))
            new_df_promo.append((0,1)[j])
            new_df_trans.append(len(df_with_date))
            try:
                new_df_negative.append(df_with_date['Плохой отзыв'].value_counts()[1]*100/len(df_with_date))
            except:
                new_df_negative.append(0)
            income=(1-sum(df_with_date['Paid'])/sum(df_with_date['Cost']))*100
            new_df_income.append(round(income,2))
```

```python
new_df['Месяц']=new_df_date
new_df['Введен промокод']=new_df_promo
new_df['Кол-во транзакций']=new_df_trans
new_df['Плохих отзывов, %']=new_df_negative
new_df['Доходность, %']=new_df_income
new_df.head()
```
